```python
import matplotlib.pyplot as plt
import numpy as np

# Data
west_scores = np.array([[8, 9, 2, 3], [5, 5, 10, 8], [7, 9, 7, 9], [8, 6, 10, 10]])
north_scores = np.array([[9, 9, 2, 3], [7, 8, 10, 8], [9, 9, 8, 9], [8, 5, 10, 10]])
riverside_scores = np.array([[8, 7, 2, 2], [7, 3, 10, 7], [4, 3, 8, 9], [1, 1, 10, 10]])

sustainability_priorities = [0.2, 0.2, 0.2, 0.4]
convenience_cost_priorities = [0.2, 0.3, 0.4, 0.1]
cost_critical_priorities = [0.1, 0.1, 0.7, 0.1]

neighborhoods = ['West', 'North', 'Riverside']
commute_types = ['Drive', 'Bus', 'Bike', 'Walk']

# Plotting bar plots for each commute type within each neighborhood
fig, axs = plt.subplots(nrows=len(neighborhoods), ncols=len(commute_types), figsize=(15, 10))

for i, neighborhood in enumerate(neighborhoods):
    for j, commute_type in enumerate(commute_types):
        scores = [west_scores, north_scores, riverside_scores][i]
        axs[i, j].bar(['Convenience', 'Speed', 'Affordability', 'Sustainability'], scores[:, j], color='skyblue')
        axs[i, j].set_title(f'{neighborhood} - {commute_type}')
        axs[i, j].set_ylabel('Score')
        axs[i, j].set_ylim(0, 12)  # Adjust the y-axis limit for better visibility

# Adding a common x-axis label for the last row
for j, commute_type in enumerate(commute_types):
    axs[len(neighborhoods) - 1, j].set_xlabel('Factors')

plt.tight_layout()
plt.show()

# Plotting student priorities
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

student_priorities = [sustainability_priorities, convenience_cost_priorities, cost_critical_priorities]
priority_labels = ['Sustainability', 'Convenience/Cost', 'Cost Critical']

for i, student_priority in enumerate(student_priorities):
    axs[i].bar(['Speed', 'Convenience', 'Affordability', 'Sustainability'], student_priority, color='lightcoral')
    axs[i].set_title(f'Student Priorities - {priority_labels[i]}')
    axs[i].set_ylabel('Weight')
    axs[i].set_ylim(0, 0.5)  # Adjust the y-axis limit for better visibility

# Adding a common x-axis label for the last subplot
axs[len(student_priorities) - 1].set_xlabel('Factors')

plt.tight_layout()
plt.show()
```
