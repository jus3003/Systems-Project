```python
import matplotlib.pyplot as plt
import numpy as np

# Data
west_scores = np.array([[8, 9, 2, 3], [5, 5, 10, 8], [7, 9, 7, 9], [8, 6, 10, 10]])
north_scores = np.array([[9, 9, 2, 3], [7, 8, 10, 8], [9, 9, 8, 9], [8, 5, 10, 10]])
riverside_scores = np.array([[8, 7, 2, 2], [7, 3, 10, 7], [4, 3, 8, 9], [1, 1, 10, 10]])

sustainability_priorities = [0.2, 0.2, 0.2, 0.4]
convenience_cost_priorities = [0.2, 0.3, 0.4, 0.1]
cost_critical_priorities = [0.1, 0.1, 0.7, 0.1]

neighborhoods = ['West', 'North', 'Riverside']
commute_types = ['Drive', 'Bus', 'Bike', 'Walk']

# Bar plots for each commute type within each neighborhood
fig, axs = plt.subplots(nrows = 3, ncols = 4, figsize=(20, 10))

for i, neighborhood in enumerate(neighborhoods):
    for j, commute_type in enumerate(commute_types):
        scores = [west_scores, north_scores, riverside_scores][i]
        axs[i, j].bar(['Convenience', 'Speed', 'Affordability', 'Sustainability'], scores[i,:], color='lightblue')
        axs[i, j].set_title(f'{neighborhood} - {commute_type}')
        axs[i, j].set_ylabel('Commute Score')
        axs[i, j].set_ylim(0, 10)

plt.tight_layout()
plt.show()

# Now plot student priorities
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize=(20, 5))

student_priorities = [sustainability_priorities, convenience_cost_priorities, cost_critical_priorities]
priority_labels = ['Sustainability', 'Convenience/Cost', 'Cost Critical']

for i, student_priority in enumerate(student_priorities):
    axs[i].bar(['Speed', 'Convenience', 'Affordability', 'Sustainability'], student_priority, color='pink')
    axs[i].set_title(f'Student Priorities - {priority_labels[i]}')
    axs[i].set_ylabel('Weight')
    axs[i].set_ylim(0, 1)  # Adjust the y-axis limit for better visibility

plt.tight_layout()
plt.show()



# Now plot city priorities
fig, axs = plt.subplots(nrows = 1, ncols = 4, figsize=(20, 5))


city_priorities = [[0.1, 0.1, 0, 0],[0.3, 0.5, 0.2, 0.1],[0.3, 0.3, 0.1, 0],[0.3, 0.3, 0, 0]]  
city_priorities = city_priorities / np.sum(city_priorities)

student_priorities = [sustainability_priorities, convenience_cost_priorities, cost_critical_priorities]
priority_labels = ['Car Infrastructure', 'Bus Infrastructure', 'Bike Infrastructure', 'Pedestrian Infrastructure']

for i, city_priority in enumerate(city_priorities):
    axs[i].bar(['Speed', 'Convenience', 'Affordability', 'Sustainability'], city_priority, color='orange')
    axs[i].set_title(f'City Priorities - {priority_labels[i]}')
    axs[i].set_ylabel('Weight')
    axs[i].set_ylim(0, 0.2)  # Adjust the y-axis limit for better visibility

plt.tight_layout()
plt.show()
```
